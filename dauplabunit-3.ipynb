```python
# (b) to test whether none of the elements of a given array is zero
import numpy as np
a=np.array([1,2,3,4])
print("\n all elements are non-zero:",np.all(a))
```

```python
#(c) to create an array wise comparsion(graeter,greathe_equal,less and less_equal,equal,equal within a tolerance) of two given arrays.
#greater
a=np.array([12,22,33])
b=np.array([45,50,55])
print("greater:",np.greater(a,b))
print("greater equal:",np.greater_equal(a,b))
print("less:",np.less(a,b))
print("less equal:",np.less_equal(a,b))
print("equal:",np.equal(a,b))
#equal within a tolerance
c=np.array([1.00001,2.00001])
d=np.array([1.00002,2.00002])
print("equal within a tolerance:",np.allclose(c,d))
```

```python
#3.2 write a numpy program using numpy methods-Max,Min,Argmax,Argmin,repr,count,bicount,unique.
import numpy as np
c = np.array([4, 7, 2, 7, 3, 4, 2,1,2,4,5,5,2, 9, 7, 2])
print("max:",np.max(c))
print("min:",np.min(c))
print("argmax:",np.argmax(c))
print("argmin:",np.argmin(c))
print("repr:",np.repr(c))
print("count:",np.count_nonzero(c))
print("bicount:",np.bincount(c))
print("unique:",np.unique(c))
```

```python
#3.2 numpy program using methods
arr=np.array([5,15,25,35,45])
#extract numbers greater than 20
print("greater than 20:",arr[arr>20])
#extract numbers less ethan 30
print("less than 30:",arr[arr<30])
#extract numbers between 20 and 40
print("between 20 and 40:",arr[(arr>20)&(arr<40)])
```

```python
#find indices of max and min numbers along a given axis
arr2d=np.array([[10,20,30],[5,25,15]])
#index of max along axis 0 (column wise)
print("Argmax(axis=0):",np.argmax(arr2d,axis=0))
#index of max along axis 1 (row wise)
print("Argmin(axis=1):",np.argmin(arr2d,axis=1))
```
